## Data Feature's
- product_id: A unique identifier for each product in the dataset.

- product_category_name: The category or type of product (e.g., electronics, clothing).

- month_year: A timestamp indicating the month and year of the transaction.

- qty: The quantity of the product sold in a given transaction.

- total_price: The total revenue generated from the sale of the product (quantity sold multiplied by unit price).

- freight_price: The shipping cost associated with the product sale.

- unit_price: The price of a single unit of the product.

- product_name_length: The length (number of characters) of the product name.

- product_description_length: The length (number of characters) of the product description.

- product_photos_qty: The number of photos available for the product.

- product_weight_g: The weight of the product in grams.

- product_score: A rating or score assigned to the product, possibly based on customer reviews.

- customers: The number of unique customers who purchased the product.

- weekday: A binary indicator for whether the transaction occurred on a weekday (1 if true, 0 if false).

- weekend: A binary indicator for whether the transaction occurred on a weekend (1 if true, 0 if false).

- holiday: A binary indicator for whether the transaction occurred on a holiday (1 if true, 0 if false).

- month: The numerical representation of the month (1-12).

- year: The year of the transaction.

- s: A placeholder or indicator that could represent a specific segment or category in the dataset (context-dependent).

- volume: The total volume of product sold, possibly factoring in size or dimensions.

- comp_1: A competitive metric or attribute related to the product, likely the first comparison point.

- ps1: Performance score or another metric associated with the first competitor.

- fp1: Another financial or performance metric related to the first competitor.

- comp_2: A second competitive metric or attribute.

- ps2: Performance score or another metric associated with the second competitor.

- fp2: A financial or performance metric related to the second competitor.

- comp_3: A third competitive metric or attribute.

- ps3: Performance score or another metric associated with the third competitor.

- fp3: A financial or performance metric related to the third competitor.

- lag_price: The price of the product in a previous period, used for trend analysis or forecasting.

### Outlier Detection and Treatment

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Outlier detection using boxplots for numerical columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Adjusting grid size based on the number of numerical columns
n_cols = 3
n_rows = (len(numerical_columns) // n_cols) + 1

# Plotting boxplots for each numerical column
plt.figure(figsize=(25, 15))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(data=data, y=col)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Outlier detection using IQR method
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Dictionary to store outlier counts for each column
outlier_counts = {}

for col in numerical_columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range
    
    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

# Display outlier counts for each column
outlier_counts_df = pd.DataFrame(list(outlier_counts.items()), columns=['Column', 'Outlier Count'])
print("Outlier Counts for Each Column:")
print(outlier_counts_df)
```

Some features show very low or near-zero correlation with qty, which might indicate limited predictive power for this specific target. These include:
product_name_length (0.117), product_description_length (-0.019), and product_photos_qty (0.026).
weekday, weekend, month, and year: All have correlations close to zero with qty, which might mean they don’t add much value for predicting quantity.

Suggested Features for the Model
Based on this correlation analysis, here’s a recommendation for the features to include in the model:

High-Correlation Features with Target (qty)

total_price, customers, and s: These show the highest positive correlation with qty and are highly predictive.
Competitive Features (Moderate Correlations)

comp_1, comp_2, comp_3: These competitive metrics (prices and scores) can add predictive power, as they give information about the external competitive landscape.

Freight and Weight Features

Seasonality and Event Indicators
